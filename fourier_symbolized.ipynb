```python
import numpy as np

T = 2*np.pi
w = 2*np.pi/T
N= 3
```

```python
import sympy as sp
x = sp.symbols('x')
n = sp.symbols('n', integer=True)

# f_input = sp.sin(x) + sp.cos(2*x)
f_input = sp.Piecewise(
    (sp.sin(x), (x % (2 * np.pi) <= np.pi) & (x % (2 * np.pi) >= 0)),
    (0, True)
)

f = sp.lambdify(x, f_input)
print(f'f = {f_input}')
```

```python
import matplotlib.pyplot as plt

d = 1000
x_val = np.linspace(0 - 2*T, 0 + 2*T, d)

y = [f(i) for i in x_val]

plt.plot(x_val, y)
plt.grid(linewidth=1)
plt.show()
```

```python
a0 = 1/T*sp.integrate(f_input, (x, -T/2, T/2))
print(f'a0 = {a0}')

an = 2/T*sp.integrate(f_input*sp.cos(n*w*x), (x, -T/2, T/2))
# print(f'an = {an}')

bn = 2/T*sp.integrate(f_input*sp.sin(n*w*x), (x, -T/2, T/2))
# print(f'bn = {bn}')

fourier = a0 + sp.Sum(an*sp.cos(n*w*x) + bn*sp.sin(n*w*x), (n, 1, N))
# print(f'fourier = {fourier}')

y_fourier = [fourier.subs(x, i) for i in x_val]
plt.plot(x_val, y_fourier)
plt.grid(linewidth=1)
plt.show()
```
