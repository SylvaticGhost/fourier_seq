```python
import numpy as np

# def f(x):
#     return np.sin(x) + np.cos(2*x)

def f(x):
    x = x % (2*np.pi)
    if 0 <= x <= np.pi:
        return np.sin(x)
    else:
        return 0

T = 2*np.pi

w = 2*np.pi/T
```

```python
import matplotlib.pyplot as plt

d = 1000
x = np.linspace(0 - 2*T, 0 + 2*T, d)

y = [f(i) for i in x]

plt.plot(x, y)
plt.grid(linewidth=1)
plt.show()
```

```python
import scipy.integrate as integrate

a0 = 2/T * integrate.quad(f, -T/2, T/2)[0]
print(f'a0 = {a0}')

def a(n):
    a_inner = lambda x: f(x)*np.cos(n*w*x)
    intg, _ = integrate.quad(a_inner, -T/2, T/2)
    return 2/T * intg

def b(n):
    b_inner = lambda x: f(x)*np.sin(n*w*x)
    intg, _ = integrate.quad(b_inner, -T/2, T/2)
    return 2/T * intg

def fourier(x, n = 35):
    sm = a0/2 + sum([a(i)*np.cos(i*w*x) + b(i)*np.sin(i*w*x) for i in range(1, n+1)])
    return sm

y_fourier = fourier(x, 20)
plt.plot(x, y_fourier)
plt.grid(linewidth=1)
plt.show()
```
